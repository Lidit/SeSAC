## BeautifulSoup

```python
from bs4 import BeautifulSoup as bs, BeautifulSoup

html_doc = '''
<html>
<head>
    <title>웹 페이지 제목</title>
</head>
<body>
    <div id="content">
        <h1>웹 페이지 내용</h1>
        <p>스크래핑을 위한 예시 문장입니다.</p>
    </div>
</body>
</html>
'''

soup = bs(html_doc, 'html.parser')

title = soup.find('title')
h1 = soup.find('h1')
p = soup.find('p')

print(title.text)
print(h1.text)
print(p.text)
```

# 인기검색 종목 가져오기

```python
import requests
from bs4 import BeautifulSoup as bs

# URL 접속
response = requests.get('https://finance.naver.com/')
print(response)

HTML = response.text
soup = bs(HTML, 'html.parser')

# 정보 찾기
popular_stock_table = soup.find('div', class_='aside_area aside_popular')
# print(popular_stock_table)

popular_stock = popular_stock_table.find_all('tr', class_=['up','down','same'])
print(popular_stock[0])

stock_list = []
for i in range(len(popular_stock)):
    stock_name = popular_stock[i].find('a').text
    stock_list.append(stock_name)
    
print(stock_list)
```

# 기상특보 가져오기

```python
response = requests.get('https://www.weather.go.kr/w/community/knowledge/standard.do')
html = response.text
soup = bs(html,'html.parser')

col = soup.find_all(scope='col')
col_list = []
for i in col:
    col_list.append(i.text)
    
# print([i.text for i in col])

row = soup.find_all(scope='row')
row_list = []
for i in row:
    row_list.append(i.text)

print(col_list)
print(row_list)

table_contents = soup.find_all('td')
table_contents_list = []
for i in table_contents:
    text = i.text
    text = text.replace('\n', '')
    text = text.replace('\t', '')
    table_contents_list.append(text)
    
# print(table_contents_list)
for i in table_contents_list:
    print(i)
#body > div.container > section > div > div.cont-wrap > div > div > table > tbody > tr:nth-child(1)
```

```python
test = soup.select('body > div.container > section > div > div.cont-wrap > div > div > table > tbody > tr')

test_dict = {}
for i in test:
    # print(i.find('th').text)
    test_dict[i.find_next().text] = [j.text.replace('\n', '').replace('\t','') for j in i.find_all('td')]

print(test_dict)
```

## 네이버 금융 배당페이지

```python
naver_finance = 'https://finance.naver.com/sise/dividend_list.naver'

html = requests.get(naver_finance).text

soup = bs(html, 'html.parser')
```

```python
# 칼럼 가져오기

columns_select = soup.select_one('#contentarea_left > table.type_1.tb_ty > thead')

columns_list = [i.text for i in columns_select.find_all('th')]

columns_list.remove('과거 3년 배당금')
print(columns_list)
```

```python
stock_data_select = soup.select_one('#contentarea_left > table.type_1.tb_ty > tbody')
stock_list = []
for i in stock_data_select.select('tr'):
    if i.select_one('.txt').text != '\xa0':
        title = i.select_one('.txt').text
        values = [j.text for j in i.select('.num')]
        result = [title] + values
        stock_list.append(result)

print(stock_list)
```

```python
import pandas as pd

df = pd.DataFrame()
df[columns_list] = columns_list

for i in stock_list:
    df = df.append(pd.Series(i, index=df.columns), ignore_index=True)

df
```

```python
#contentarea_left > table.Nnavi > tbody > tr > td.pgRR > a
last_page = int(soup.select_one('td.pgRR').find_next().get('href').split("&page=")[-1])
print(last_page)
```

```python
def get_page_data(page_num):
    naver_finance = 'https://finance.naver.com/sise/dividend_list.naver'
    html = requests.get(naver_finance+f'?&page={str(page_num)}').text
    # print(html)
    soup = bs(html, 'html.parser')
    # 칼럼 가져오기

    columns_select = soup.select_one('#contentarea_left > table.type_1.tb_ty > thead')

    columns_list = [i.text for i in columns_select.find_all('th')]

    columns_list.remove('과거 3년 배당금')
    stock_data_select = soup.select_one('#contentarea_left > table.type_1.tb_ty > tbody')
    stock_data = []
    for i in stock_data_select.select('tr'):
        if i.select_one('.txt').text != '\xa0':
            title = i.select_one('.txt').text
            values = [j.text for j in i.select('.num')]
            result = [title] + values
            stock_data.append(result)
    df = pd.DataFrame(stock_data, columns=columns_list)
    
    return df
```

```python
empty_df = pd.DataFrame()
for i in range(1,last_page+1):
    empty_df = pd.concat([empty_df, get_page_data(i)])
empty_df = empty_df.reset_index(drop=True)
empty_df
```
