### 4-2. 변수 선택2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

from tensorflow import keras
from keras.layers import Dense
from keras.models import Sequential, load_model

# 샘플 데이터 생성
# np.random.seed(0)

# 날짜 범위 생성
# date_range = pd.date_range(start='2010-01-01', end='2022-12-31', freq='MS')
date_range = pd.date_range(start='2010-01-01', end='2022-12-31', freq='M')

###

# 랜덤 다변량 데이터 생성 (환율, 주가, 금리 등)
num_samples = len(date_range)
exchange_rate = 100 + np.cumsum(np.random.randn(num_samples))
stock_price = 1000 + np.cumsum(np.random.randn(num_samples))
interest_rate = 0.02 + np.cumsum(0.001 * np.random.randn(num_samples))

print('date_range >> ', date_range)
print('num_smaples >> ', num_samples)
print('exchange_rate : >> ', exchange_rate.shape)
print('stock_price ', stock_price.shape)
print('interest_rate : >> ', interest_rate.shape)
```

```python

# 데이터프레임 생성
data = pd.DataFrame({'Date': date_range, 'ExchangeRate': exchange_rate, 'StockPrice': stock_price, 'InterestRate': interest_rate})
data.set_index('Date', inplace=True)

# 데이터 시각화
plt.figure(figsize=(12, 6))
plt.plot(data['ExchangeRate'], label='Exchange Rate')
plt.plot(data['StockPrice'], label='Stock Price')
plt.plot(data['InterestRate'], label='Interest Rate')
plt.legend()
plt.title('다변량 데이터 시각화')
plt.show()

# 데이터 정규화
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# 학습 데이터와 테스트 데이터 분리
train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

# 시계열 데이터를 입력 시퀀스로 변환
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30  # 시퀀스 길이 설정
X_train, y_train = create_sequences(train_data, seq_length)
X_test, y_test = create_sequences(test_data, seq_length)

# LSTM 모델 생성
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(seq_length, 3)))
model.add(Dense(3))  # 출력 노드 수는 다변량 데이터의 변수 개수와 동일해야 함

# 모델 컴파일
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

########
# 모델 학습
model.fit(X_train, y_train, epochs=100, batch_size=64)

# 모델 평가
test_loss = model.evaluate(X_test, y_test)
print(f"테스트 데이터 손실: {test_loss:.4f}")

# 예측
y_pred = model.predict(X_test)

# 정규화된 데이터를 원래 스케일로 변환
y_test = scaler.inverse_transform(y_test)
y_pred = scaler.inverse_transform(y_pred)

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index[-len(y_test):], y_test[:, 0], label='실제 환율')
plt.plot(data.index[-len(y_pred):], y_pred[:, 0], label='예측 환율')
plt.legend()
plt.title('다변량 환율 예측 결과')
plt.show()
```
