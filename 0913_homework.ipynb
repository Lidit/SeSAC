```python
import pandas as pd
mpg = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/mpg.csv')
mpg
```

# 1. 자동차 제조사에 따른 자동차 종류의 개수 시각화


```python
import matplotlib.pyplot as plt
import numpy as np
df = mpg[['manufacturer','model']].groupby('manufacturer').count()

x = list(df.index)
# df.reset_index()
y = df.model.values

fig = plt.figure(figsize=(30,10))

axes1 = fig.add_subplot(2,2,1) 
axes1.bar(x,y)
```
