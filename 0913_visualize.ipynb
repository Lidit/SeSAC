# matplotlib

```python
# !pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100, 300, 10)
y = np.random.randn(20)

plt.plot(x,y)
```

```python
plt.figure(figsize = (15,5))
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
```

```python
plt.plot(x, y, color = 'r', linestyle = '--', marker = 's')
```

#   

```python
fig = plt.figure(figsize=(16,10))

axes1 = fig.add_subplot(2,2,1) # 2x2   
axes1.plot(x, y, color = 'r', linestyle = '--', marker = 's')

axes2 = fig.add_subplot(2,2,2) # 2x2   
axes2.hist(y) #

axes2 = fig.add_subplot(2,2,3) # 2x2   
axes2.scatter(x,y) #
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,10))
axes[0,0].plot(x,y, color='skyblue')
axes[0,1].hist(y)
axes[1,0].scatter(x,y)
axes[1,1].plot(x,y,'r--s')
```

## 2 graph in 1 figure

```python
x1 = np.arange(100,300,10)
y1 = np.random.randn(20)

x2 = np.arange(200,400,10)
y2 = np.random.randn(20)

plt.plot(x1,y1, label='x1')
plt.plot(x2,y2, label='x2')

plt.legend(loc=1) #  (loc: )
```

##   label, title 

```python
fig = plt.figure(figsize=(16,10))

axes1 = fig.add_subplot(2,2,1) # 2x2   
axes1.plot(x, y, color = 'r', linestyle = '--', marker = 's')
axes1.set_title('axes1_title')

axes2 = fig.add_subplot(2,2,2) # 2x2   
axes2.hist(y) #
axes2.set_title('axes2_title')

axes3 = fig.add_subplot(2,2,3) # 2x2   
axes3.scatter(x,y) #
axes3.set_title('axes3_title')

fig.suptitle('Figure Title')
```
